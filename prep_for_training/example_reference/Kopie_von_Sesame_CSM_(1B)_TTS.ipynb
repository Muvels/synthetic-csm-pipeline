### Data Prep  

We will use the `MrDragonFox/Elise`, which is designed for training TTS models. Ensure that your dataset follows the required format: **text, audio** for single-speaker models or **source, text, audio** for multi-speaker models. You can modify this section to accommodate your own dataset, but maintaining the correct structure is essential for optimal training.

```python
# Dataset Prep functions
import os
import re
from pathlib import Path

from datasets import Audio, Dataset
from transformers import AutoProcessor
import torch
from soundfile import LibsndfileError
import soundfile as sf
import gc

# ---------------------------
# Config
# ---------------------------
DATA_ROOT = Path("./dataset")  # <-- your dataset root
target_sampling_rate = 1500

MAX_AUDIO_SECONDS = 15              # <-- hard limit: only keep <= 15s
MAX_AUDIO_SAMPLES = target_sampling_rate * MAX_AUDIO_SECONDS  # 15s -> 360000

MAX_DATASET_SIZE = 2500             # <-- reduce if you still OOM (try 400 / 300)

processor = AutoProcessor.from_pretrained("unsloth/csm-1b")

# ---------------------------
# Build HF Dataset from local folder structure
# ---------------------------
examples = []

for uuid_dir in sorted(p for p in DATA_ROOT.iterdir() if p.is_dir()):
    segments_dir = uuid_dir / "segments"
    if not segments_dir.is_dir():
        print(f"Warning: no 'segments' folder in {uuid_dir}, skipping.")
        continue

    script_path = segments_dir / "vibevoice-podcast-script.txt"
    if not script_path.is_file():
        print(f"Warning: no script file in {segments_dir}, skipping.")
        continue

    # 1) Parse transcript file: [1]: text...
    script_lines = []
    with script_path.open("r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            m = re.match(r"\[(\d+)\]:\s*(.*)", line)
            if not m:
                print(f"Warning: could not parse line in {script_path}: {line}")
                continue
            speaker_id = m.group(1)  # "1", "2", ...
            text = m.group(2)
            script_lines.append({"speaker": speaker_id, "text": text})

    # 2) Collect wav files in segments folder
    wav_paths = sorted(p for p in segments_dir.glob("*.wav"))
    if len(wav_paths) != len(script_lines):
        print(
            f"Warning: mismatch in {uuid_dir}: "
            f"{len(wav_paths)} wavs vs {len(script_lines)} script lines. "
            f"Using min length."
        )

    for wav_path, line in zip(wav_paths, script_lines):
        examples.append(
            {
                "audio": str(wav_path),  # HF Audio feature will load this
                "text": line["text"],
                "source": line["speaker"],  # column used as speaker id
            }
        )

print(f"Total examples before any filtering: {len(examples)}")

# Create HF Dataset from list and cast audio column
raw_ds = Dataset.from_list(examples)
# After this, example["audio"] becomes a dict: {"path": ..., "array": np.array}
raw_ds = raw_ds.cast_column("audio", Audio(sampling_rate=target_sampling_rate))

# We don't need the Python list anymore
del examples
gc.collect()

# ---------------------------
# Filter out audio > MAX_AUDIO_SECONDS and corrupted files
# ---------------------------
def filter_long_audio(example):
    try:
        audio_path = example["audio"]["path"]
        audio_info = sf.info(audio_path)
        num_samples = audio_info.frames
        return num_samples <= MAX_AUDIO_SAMPLES
    except LibsndfileError as e:
        print(f"Warning: Skipping audio file '{example['audio']['path']}' due to LibsndfileError: {e}")
        return False
    except FileNotFoundError:
        print(f"Warning: Skipping audio file '{example['audio']['path']}' as it was not found.")
        return False
    except Exception as e:
        print(f"Warning: Skipping audio file '{example['audio']['path']}' due to unexpected error: {e}")
        return False

raw_ds = raw_ds.filter(
    filter_long_audio,
    desc=f"Filtering audio > {MAX_AUDIO_SECONDS}s and corrupted files",
)

print(f"After length/filter: {len(raw_ds)} examples (<= {MAX_AUDIO_SECONDS}s only)")

# Limit dataset size to avoid out-of-memory
if len(raw_ds) > MAX_DATASET_SIZE:
    raw_ds = raw_ds.select(range(MAX_DATASET_SIZE))
    print(f"Reduced dataset size to {len(raw_ds)} examples (cap = {MAX_DATASET_SIZE}).")

# ---------------------------
# Speaker column logic
# ---------------------------
speaker_key = "source"
if "source" not in raw_ds.column_names and "speaker_id" not in raw_ds.column_names:
    print('Unsloth: No speaker found, adding default "source" of 0 for all examples')
    new_column = ["0"] * len(raw_ds)
    raw_ds = raw_ds.add_column("source", new_column)
elif "source" not in raw_ds.column_names and "speaker_id" in raw_ds.column_names:
    speaker_key = "speaker_id"

# ---------------------------
# Preprocess function
# ---------------------------
def preprocess_example(example):
    # All long/corrupted audios should already be filtered out
    audio_array = example["audio"]["array"]

    conversation = [
        {
            "role": str(example[speaker_key]),
            "content": [
                {"type": "text", "text": example["text"]},
                {"type": "audio", "path": audio_array},
            ],
        }
    ]

    model_inputs = processor.apply_chat_template(
        conversation,
        tokenize=True,
        return_dict=True,
        output_labels=True,
        text_kwargs={
            "padding": "max_length",
            "max_length": 256,        # text tokens
            "pad_to_multiple_of": 8,
            "padding_side": "right",
        },
        audio_kwargs={
            "sampling_rate": target_sampling_rate,
            "max_length": MAX_AUDIO_SAMPLES,   # 15s -> 360000 samples
            "padding": "max_length",
        },
        common_kwargs={"return_tensors": "pt"},
    )

    required_keys = [
        "input_ids",
        "attention_mask",
        "labels",
        "input_values",
        "input_values_cutoffs",
    ]
    processed_example = {}

    for key in required_keys:
        if key not in model_inputs:
            raise ValueError(
                f"Required key '{key}' not in processor output. "
                f"Got keys: {list(model_inputs.keys())}"
            )
        processed_example[key] = model_inputs[key][0]  # first (and only) batch elem

    # Safety check: ensure all tensors
    if not all(isinstance(v, torch.Tensor) for v in processed_example.values()):
        print(
            "Error: Not all required keys are tensors in final processed example. "
            f"Types: { {k: type(v) for k, v in processed_example.items()} }"
        )

    return processed_example

# ---------------------------
# Map over dataset (single pass, no extra filter afterwards)
# ---------------------------
processed_ds = raw_ds.map(
    preprocess_example,
    remove_columns=raw_ds.column_names,  # drop text/audio/source from final dataset
    desc="Preprocessing dataset",
    # num_proc=1 by default ‚Äî keep it 1 to save RAM
)

print("Finished preprocessing.")
print(processed_ds)

# Copy some samples of raw_ds for infernece
raw_ds_inference = raw_ds[:20]

# Free raw_ds to recover RAM before training
del raw_ds
gc.collect()
```

```python
# Find max raw audio length in samples
max_audio_len = 0

for ex in raw_ds_inference:
    # ex["audio"]["array"] is a 1D float32 array
    cur_len = len(ex["audio"]["array"])
    if cur_len > max_audio_len:
        max_audio_len = cur_len

print("Max raw audio length (samples):", max_audio_len)
print("Max raw audio length (seconds):", max_audio_len / target_sampling_rate)
```

<a name="Train"></a>
### Train the model
Now let's use Huggingface  `Trainer`! More docs here: [Transformers docs](https://huggingface.co/docs/transformers/main_classes/trainer). We do 60 steps to speed things up, but you can set `num_train_epochs=1` for a full run, and turn off `max_steps=None`.

```python
from transformers import TrainingArguments, Trainer
from unsloth import is_bfloat16_supported

trainer = Trainer(
    model = model,
    train_dataset = processed_ds,
    args = TrainingArguments(
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4,
        warmup_steps = 5,
        max_steps = 60,
        learning_rate = 2e-4,
        fp16 = not is_bfloat16_supported(),
        bf16 = is_bfloat16_supported(),
        logging_steps = 1,
        optim = "adamw_8bit",
        weight_decay = 0.001, # Turn this on if overfitting
        lr_scheduler_type = "linear",
        seed = 3407,
        output_dir = "outputs",
        report_to = "none", # Use TrackIO/WandB etc
    ),
)
```

```python
# @title Show current memory stats
gpu_stats = torch.cuda.get_device_properties(0)
start_gpu_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
max_memory = round(gpu_stats.total_memory / 1024 / 1024 / 1024, 3)
print(f"GPU = {gpu_stats.name}. Max memory = {max_memory} GB.")
print(f"{start_gpu_memory} GB of memory reserved.")
```

```python
trainer_stats = trainer.train()
```

```python
# @title Show final memory and time stats
used_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
used_memory_for_lora = round(used_memory - start_gpu_memory, 3)
used_percentage = round(used_memory / max_memory * 100, 3)
lora_percentage = round(used_memory_for_lora / max_memory * 100, 3)
print(f"{trainer_stats.metrics['train_runtime']} seconds used for training.")
print(
    f"{round(trainer_stats.metrics['train_runtime']/60, 2)} minutes used for training."
)
print(f"Peak reserved memory = {used_memory} GB.")
print(f"Peak reserved memory for training = {used_memory_for_lora} GB.")
print(f"Peak reserved memory % of max memory = {used_percentage} %.")
print(f"Peak reserved memory for training % of max memory = {lora_percentage} %.")
```

```python
speaker_id = 0

utterance = raw_ds_inference["audio"][5]["array"]
utterance_text = raw_ds_inference["text"][5]
text = "Guten Tag, Ich bin ihre Virtuelle Assistenz. Wie kann ich ihnen heute helfen?"

# CSM will fill in the audio for the last text.
# You can even provide a conversation history back in as you generate new audio

conversation = [
    {"role": str(2), "content": [{"type": "text", "text": raw_ds_inference["text"][3]},{"type": "audio", "path": raw_ds_inference["audio"][3]["array"]}]},
    {"role": str(speaker_id), "content": [{"type": "text", "text": utterance_text},{"type": "audio", "path": utterance}]},
    {"role": str(speaker_id), "content": [{"type": "text", "text": text}]},
]

inputs = processor.apply_chat_template(
        conversation,
        tokenize=True,
        return_dict=True,
    )
audio_values = model.generate(
    **inputs.to("cuda"),
    max_new_tokens=125, # 125 tokens is 10 seconds of audio, for longer text increase this
    # play with these parameters to tweak results
    # depth_decoder_top_k=0,
    # depth_decoder_top_p=0.9,
    # depth_decoder_do_sample=True,
    # depth_decoder_temperature=0.9,
    # top_k=0,
    # top_p=1.0,
    # temperature=0.9,
    # do_sample=True,
    #########################################################
    output_audio=True
)
audio = audio_values[0].to(torch.float32).cpu().numpy()
sf.write("example_with_context.wav", audio, 24000)
display(Audio(audio, rate=24000))
```
